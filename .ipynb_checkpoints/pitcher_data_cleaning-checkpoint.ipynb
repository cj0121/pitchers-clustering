## Group pitches
I'll group similar pitches to reduce variability and drop some of the weird ones that rarely occur. 

```python
data_train[['pitch_type', 'pitch_name']].value_counts()
```

```python
pitch_group = {'CH':'off_speed', 'FF':'fastball', 'SL':'sli_cut', 'CU':'curve', 'KC':'curve', 
               'SI':'fastball', 'FC':'sli_cut', 'CS':'curve', 'FS':'off_speed', 'FA':'other', 
               'EP':'other', 'KN':'other', 'SC':'other', 'FO':'other'}
```

```python
pitch_group2 = {'CH':'off_speed', 'FF':'fastball', 'SL':'breaking', 'CU':'breaking', 'KC':'breaking', 
               'SI':'fastball', 'FC':'fastball', 'CS':'breaking', 'FS':'off_speed', 'FA':'other', 
               'EP':'other', 'KN':'other', 'SC':'other', 'FO':'other'}
```

```python
# add group label
for index, row in data_train.iterrows():
    data_train.at[index, 'pitch_group'] = pitch_group[row.pitch_type]
    
for index, row in data_train.iterrows():
    data_train.at[index, 'pitch_group2'] = pitch_group2[row.pitch_type]
# drop "others". This applies to group2 as well
data_train = data_train[data_train['pitch_group']!='other'].copy()
```

## 2.1 Player Profile
### Handedness

```python
# add handedness label
data_train['throws_right'] = 0
data_train.loc[data_train['p_throws']=='R', 'throws_right'] = 1
```

### Pitch Arsenal

```python
# create dummies for pitch groups
# group 1
dummies = pd.get_dummies(data_train['pitch_group'])
data_train = pd.concat((data_train, dummies), axis=1)
# group 2
dummies2 = pd.get_dummies(data_train['pitch_group2'])
dummies2.rename(columns={'fastball':'fastball_2', 'breaking':'breaking_2', 'off_speed':'off_speed_2'}, inplace=True)
data_train = pd.concat((data_train, dummies2), axis=1)

data_train['total_pitch'] = 1 # pitch count to calculate percentages
```

```python
# group by pitcher
data_player_full = data_train.groupby(['pitcher']).agg({'player_name':max, 'throws_right':max,
                                                         'release_pos_x':'mean', 'release_pos_z':'mean', 
                                                         'fastball':sum, 'curve':sum, 'off_speed':sum,
                                                         'sli_cut':sum, 'fastball_2':sum, 'breaking_2':sum,
                                                         'off_speed_2':sum, 'total_pitch':sum}).reset_index().copy()
```

```python
# calculate percentages
# group 1
data_player_full['fastball_pct'] = data_player_full['fastball']/data_player_full['total_pitch']
data_player_full['curve_pct'] = data_player_full['curve']/data_player_full['total_pitch']
data_player_full['off_speed_pct'] = data_player_full['off_speed']/data_player_full['total_pitch']
data_player_full['sli_cut_pct'] = data_player_full['sli_cut']/data_player_full['total_pitch']
# group 2
data_player_full['fastball_2pct'] = data_player_full['fastball_2']/data_player_full['total_pitch']
data_player_full['breaking_2pct'] = data_player_full['breaking_2']/data_player_full['total_pitch']
data_player_full['off_speed_2pct'] = data_player_full['off_speed_2']/data_player_full['total_pitch']
```

```python
print(data_player_full.shape)
data_player_full.head()
```

```python
# seperate data based on grouping methods
data_player1 = data_player_full[['pitcher', 'player_name', 'throws_right', 'release_pos_x',
       'release_pos_z', 'fastball_pct', 'curve_pct', 'off_speed_pct', 'sli_cut_pct']].copy()
data_player2 = data_player_full[['pitcher', 'player_name', 'throws_right', 'release_pos_x',
       'release_pos_z', 'fastball_2pct', 'breaking_2pct', 'off_speed_2pct']].copy()
```

```python
data_player2[data_player2.fastball_2pct==0].head() # see who doesn't throw fastball
```

Many of them were just position players brought in to eat through the innings in hopeless situations. They threw a lot of "FA" - slow fastballs, which were all dropped.

```python
# for group 2, keep only pitchers who throw all three pitch types
data_player2_drop = data_player2[(data_player2.fastball_2pct!=0)&(data_player2.breaking_2pct!=0)&
                                 (data_player2.off_speed_2pct!=0)].copy()
data_player2_drop.shape
```

## 2.3 Contact Profile

```python
data_contact = data_train_full.copy() # start with full dataset
print(data_contact.shape)
data_contact.head()
```

```python
# check and drop all NAN because only balls in play have launch info
var_check = ['launch_speed', 'launch_angle', 'launch_speed_angle']
data_contact[var_check].isna().sum()
```

```python
# drop all nan
data_contact = data_contact.dropna(subset=var_check, how='any').copy()
data_contact.shape
```

### Barrel + Solid Percentage

```python
# create total count
data_contact['is_inPlay'] = 1

# label data if they are solid contact: barrel + solid
data_contact['is_solidContact'] = 0
data_contact.loc[(data_contact['launch_speed_angle']==5)|(data_contact['launch_speed_angle']==6), 'is_solidContact'] = 1
```

### Launch Angle Rate and Average Launch Speed

```python
data_contact.bb_type.unique()
```

```python
# drop NAN on 'bb_type'
data_contact = data_contact[data_contact.bb_type.notna()].copy()
data_contact.shape # 1 row dropped
```

```python
# create dummies on bb_type
dummies = pd.get_dummies(data_contact['bb_type'])
data_contact = pd.concat((data_contact, dummies), axis=1)
```

```python
dummies.columns
```

### Group data by pitcher and calculate rates

```python
# group data by pitcher and calculate rates
data_contact_group = data_contact.groupby(['pitcher']).agg({'launch_speed':'mean', 'is_solidContact':sum, 
                                                              'is_inPlay':sum, 'ground_ball':sum, 'line_drive':sum,
                                                             'fly_ball':sum, 'popup':sum}).copy()

data_contact_group['solid_rate'] = data_contact_group['is_solidContact']/data_contact_group['is_inPlay']
data_contact_group['GB_rate'] = data_contact_group['ground_ball']/data_contact_group['is_inPlay']
data_contact_group['LD_rate'] = data_contact_group['line_drive']/data_contact_group['is_inPlay']
data_contact_group['FB_rate'] = data_contact_group['fly_ball']/data_contact_group['is_inPlay']
data_contact_group['PU_rate'] = data_contact_group['popup']/data_contact_group['is_inPlay']
```

```python
data_contact_group = data_contact_group.reset_index().copy()
data_contact_group = data_contact_group[['pitcher', 'launch_speed', 'solid_rate', 
                                             'GB_rate', 'LD_rate','FB_rate', 'PU_rate']].copy()
```

```python
print(data_contact_group.shape)
data_contact_group.head()
```

## 2.4 Swing Profile

```python
# get full dataset
data_swing = data_train_full.copy()
```

```python
data_swing['description'].unique()
```

### Whiff Rate

```python
# label pitch if it is a swing or swing and miss
is_swing = ['swinging_strike', 'foul', 'hit_into_play', 'foul_tip', 'swinging_strike_blocked']
data_swing['is_swing'] = 0
data_swing.loc[data_swing['description'].isin(is_swing), 'is_swing'] = 1
data_swing['is_whiff'] = 0
data_swing.loc[data_swing['description']=='swinging_strike', 'is_whiff'] = 1
```

### O-Swing%

Check if data has missing value on where a pitch crosses the plate, as well as the strike zone vertical boundaries since they vary based on the batter

```python
# check and drop NA
check_var = ['plate_x', 'plate_z', 'sz_top', 'sz_bot']
data_swing[check_var].isna().sum()
```

```python
data_swing.shape
```

```python
# drop NA
data_swing = data_swing.dropna(subset=check_var, how='any').copy()
data_swing.shape
```

```python
# define strikezone size
szX_min = -10/12 # half of home plate + half of baseball
szX_max = 10/12
data_swing['sz_topTrue'] = data_swing['sz_top'] + 1.5/12 # add half a baseball
data_swing['sz_botTrue'] = data_swing['sz_bot'] - 1.5/12

# label outside of strikezone pitches
data_swing['is_outside'] = 1
data_swing.loc[(data_swing['plate_x']>=szX_min)&
               (data_swing['plate_x']<=szX_max)&
               (data_swing['plate_z']>=data_swing['sz_botTrue'])&
               (data_swing['plate_z']<=data_swing['sz_topTrue']), 'is_outside'] = 0

# label outside swing
data_swing['is_OSwing'] = 0
data_swing.loc[(data_swing['is_swing']==1)&(data_swing['is_outside']==1), 'is_OSwing'] = 1
```

### Group Data

```python
# group data
data_swing_group = data_swing.groupby(['pitcher']).agg({'is_swing':sum, 'is_whiff':sum, 
                                                          'is_outside':sum, 'is_OSwing':sum}).reset_index().copy()
```

```python
# calculate whiff_rate and OSwing_pct
data_swing_group['whiff_rate'] = data_swing_group['is_whiff']/data_swing_group['is_swing']
data_swing_group['OSwing_pct'] = data_swing_group['is_OSwing']/data_swing_group['is_outside']

data_swing_group = data_swing_group[['pitcher', 'whiff_rate', 'OSwing_pct']].copy()
```

```python
data_swing_group.shape
```

```python
data_swing_group.head()
```

## 2.5 Production Stats
K/9, Hit/9, BB/9, HR/9

```python
# get full dataset
data_stats = data_train_full.copy()
# group data by plate appearance
data_stats_pa = data_stats.groupby(['game_date', 
                                    'game_pk', 'at_bat_number']).agg({'game_year':max,
                                                                    'home_team':max,
                                                                    'away_team':max,
                                                                    'player_name':max,
                                                                    'batter':max,
                                                                    'pitcher':max,
                                                                    'inning_topbot':max,
                                                                    'inning':max,
                                                                    'events':'first'}).reset_index().copy()
```

```python
# check and drop NA on 'events'
print(f"Number of NA: {data_stats_pa['events'].isna().sum()}")
```

```python
data_stats_pa = data_stats_pa[data_stats_pa['events'].notna()].copy() # drop NA
```

### Calculate individual stats and IP per pitcher

```python
data_stats_pa['events'].unique()
```

#### Stats

```python
is_K = ['strikeout', 'strikeout_double_play']
is_hit = ['single', 'home_run', 'double', 'triple']
is_BB = ['walk', 'hit_by_pitch']
is_HR = ['home_run']

data_stats_pa['is_K'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_K), 'is_K'] = 1

data_stats_pa['is_hit'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_hit), 'is_hit'] = 1

data_stats_pa['is_BB'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_BB), 'is_BB'] = 1

data_stats_pa['is_HR'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_HR), 'is_HR'] = 1

# data_stats_pa['PA_pitched'] = 1
```

#### Outs

```python
is_1out = ['field_out', 'strikeout', 'fielders_choice', 'sac_bunt', 'sac_fly', 'force_out',
          'fielders_choice_out', 'caught_stealing_2b', 'caught_stealing_home', 'pickoff_1b',
          'other_out', 'caught_stealing_3b', 'pickoff_2b', 'pickoff_caught_stealing_home',
          'pickoff_caught_stealing_3b', 'pickoff_3b', 'pickoff_caught_stealing_2b']
is_2out = ['grounded_into_double_play', 'double_play', 'strikeout_double_play',
          'sac_bunt_double_play', 'sac_fly_double_play', 'runner_double_play']
is_3out = ['triple_play']
is_0out = ['single', 'home_run', 'hit_by_pitch', 'walk', 'double', 'field_error', 'triple',
           'wild_pitch', 'catcher_interf', 'game_advisory', 'passed_ball', 'stolen_base_2b']

data_stats_pa['num_outs'] = 3
data_stats_pa.loc[data_stats_pa.events.isin(is_1out), 'num_outs'] = 1
data_stats_pa.loc[data_stats_pa.events.isin(is_2out), 'num_outs'] = 2
data_stats_pa.loc[data_stats_pa.events.isin(is_0out), 'num_outs'] = 0
```

```python
# group data and calculate stats
data_stats_group = data_stats_pa.groupby(['pitcher']).agg({'is_K':sum, 'is_hit':sum, 
                                                            'is_BB':sum, 'is_HR':sum,
                                                            'num_outs':sum}).reset_index().copy()
data_stats_group['IP'] = data_stats_group['num_outs']/3
data_stats_group['K_9'] = (data_stats_group['is_K']/data_stats_group['IP'])*9
data_stats_group['hit_9'] = (data_stats_group['is_hit']/data_stats_group['IP'])*9
data_stats_group['BB_9'] = (data_stats_group['is_BB']/data_stats_group['IP'])*9
data_stats_group['HR_9'] = (data_stats_group['is_HR']/data_stats_group['IP'])*9
```

```python
# save a copy of full columns
data_stats_group_full = data_stats_group.copy()
```

```python
data_stats_group = data_stats_group_full[['pitcher', 'IP','K_9', 'hit_9', 'BB_9', 'HR_9']].copy()
```

# 5. Neutralize handedness
A lot of these pitch data contains handedness information in a numerical form (i.e. +/- release_pos_x). Most clustering algorithms supposedly can't handle such information, however I did not verify this. In this case, I will force all pitchers to be right-handed.

```python
data_neutral = data_drop.copy()
```

Drop spin axis on breaking balls as curveballs have opposite spin axis than all other pitches. Otherwise, pitchers' breaking ball spin axis will greatly vary depending on how many curveballs they throw

```python
data_neutral.drop(columns=['spin_axis_breaking', 'IP'], inplace=True) # drop IP as well to make data cleaner
```

```python
data_neutral['release_pos_x'] = abs(data_neutral['release_pos_x'])*-1 # make all release_pos_x negative
```

```python
# get reversed spin axis on lefties
data_neutral.loc[data_neutral['throws_right']==0, 'spin_axis_fastball'] = 360 - data_neutral.loc[data_neutral['throws_right']==0, 'spin_axis_fastball']
data_neutral.loc[data_neutral['throws_right']==0, 'spin_axis_off_speed'] = 360 - data_neutral.loc[data_neutral['throws_right']==0, 'spin_axis_off_speed']
```

```python
data_neutral.head()
```

Neutralization was not performed on pitch movement data as it should be. Statcast's pitch movement data is complicated, more investigation needs to be done before pfx can be neutralized.
