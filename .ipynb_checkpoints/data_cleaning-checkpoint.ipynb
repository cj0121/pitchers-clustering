Group By:

Fastball: FF

Moving Fastball: SI

Slider/Cutter: SL, FC

Curve: CU, KC, CS

Off-speed: CH, FS, FO

other: FA, KN, EP, SC

```python
pitch_group = {'FF':'fastball', 'SI':'fastball', 'SL':'sli_cut', 'FC':'sli_cut', 'CU':'curve',
              'KC':'curve', 'CS':'curve', 'CH':'off_speed', 'FS':'off_speed', 'FO':'off_speed',
              'FA':'other', 'KN':'other', 'EP':'other', 'SC':'other'}
```

```python
pitch_group2 = {'FF':'fastball', 'SI':'fastball', 'SL':'breaking', 'FC':'fastball', 'CU':'breaking',
              'KC':'breaking', 'CS':'breaking', 'CH':'off_speed', 'FS':'off_speed', 'FO':'off_speed',
              'FA':'other', 'KN':'other', 'EP':'other', 'SC':'other'}
```

```python
# add group label
for index, row in data_train.iterrows():
    data_train.at[index, 'pitch_group'] = pitch_group[row.pitch_type]
    
for index, row in data_train.iterrows():
    data_train.at[index, 'pitch_group2'] = pitch_group2[row.pitch_type]
```

```python
# drop "others"
data_train = data_train[data_train['pitch_group']!='other'].copy()
```

```python
data_contact.head()
```

```python
# drop all nan
data_contact = data_contact.dropna(subset=var_check, how='any').copy()
data_contact.shape
```

```python
# # remove all pitch that start with 2 strkes
# data_contact = data_contact[data_contact['strikes']!=2].copy()
# data_contact.shape
```

### get barrel and solid percent of all balls in play

```python
# add hit_into_play label

data_contact['is_inPlay'] = 0
data_contact.loc[data_contact['description']=='hit_into_play', 'is_inPlay'] = 1
data_contact['is_solidContact'] = 0
data_contact.loc[(data_contact['launch_speed_angle']==5)|(data_contact['launch_speed_angle']==6), 'is_solidContact'] = 1
```

### label launch angle

```python
data_contact['is_GB'] = 0
data_contact['is_LD'] = 0
data_contact['is_FB'] = 0
data_contact['is_PU'] = 0
data_contact.loc[data_contact['launch_angle']<10, 'is_GB'] = 1
data_contact.loc[(data_contact['launch_angle']>=10)&(data_contact['launch_angle']<25), 'is_LD'] = 1
data_contact.loc[(data_contact['launch_angle']>=25)&(data_contact['launch_angle']<50), 'is_FB'] = 1
data_contact.loc[data_contact['launch_angle']>50, 'is_PU'] = 1
```

```python
# get average launch speed
data_contact_grouped = data_contact.groupby(['pitcher']).agg({'launch_speed':'mean', 'is_solidContact':sum, 
                                                              'is_inPlay':sum, 'is_GB':sum, 'is_LD':sum,
                                                             'is_FB':sum, 'is_PU':sum})
```

```python
data_contact_grouped['solid_rate'] = data_contact_grouped['is_solidContact']/data_contact_grouped['is_inPlay']
data_contact_grouped['GB_rate'] = data_contact_grouped['is_GB']/data_contact_grouped['is_inPlay']
data_contact_grouped['LD_rate'] = data_contact_grouped['is_LD']/data_contact_grouped['is_inPlay']
data_contact_grouped['FB_rate'] = data_contact_grouped['is_FB']/data_contact_grouped['is_inPlay']
data_contact_grouped['PU_rate'] = data_contact_grouped['is_PU']/data_contact_grouped['is_inPlay']
```

```python
data_contact_grouped = data_contact_grouped.reset_index()
data_contact_grouped = data_contact_grouped[['pitcher', 'launch_speed', 'solid_rate', 
                                             'GB_rate', 'LD_rate','FB_rate', 'PU_rate']].copy()
```

# Swing Profile
### whiff rate

```python
# get full dataset
data_swing = data_train_full.copy()
```

```python
data_swing['description'].unique()
```

```python
# label is_swing
is_swing = ['swinging_strike', 'foul', 'hit_into_play', 'foul_tip', 'swinging_strike_blocked']
data_swing['is_swing'] = 0
data_swing.loc[data_swing['description'].isin(is_swing), 'is_swing'] = 1
data_swing['is_whiff'] = 0
data_swing.loc[data_swing['description']=='swinging_strike', 'is_whiff'] = 1
```

### O Swing%

```python
# check NA
check_var = ['plate_x', 'plate_z', 'sz_top', 'sz_bot']
data_swing[check_var].isna().sum()
```

```python
# drop NA
data_swing = data_swing.dropna(subset=check_var, how='any').copy()
data_swing.shape
```

```python
# define strikezone size
szX_min = -10/12 # half of home plate + half of baseball
szX_max = 10/12
data_swing['sz_topTrue'] = data_swing['sz_top'] + 1.5/12
data_swing['sz_botTrue'] = data_swing['sz_bot'] - 1.5/12

# label outside of strikezone pitches
data_swing['is_outside'] = 1
data_swing.loc[(data_swing['plate_x']>=szX_min)&(data_swing['plate_x']<=szX_max)&(data_swing['plate_z']>=data_swing['sz_botTrue'])&(data_swing['plate_z']<=data_swing['sz_topTrue']), 'is_outside'] = 0

# label outside swing
data_swing['is_OSwing'] = 0
data_swing.loc[(data_swing['is_swing']==1)&(data_swing['is_outside']==1), 'is_OSwing'] = 1
```

### Group Data

```python
# group data
data_swing_grouped = data_swing.groupby(['pitcher']).agg({'is_swing':sum, 'is_whiff':sum, 
                                                          'is_outside':sum, 'is_OSwing':sum})
data_swing_grouped = data_swing_grouped.reset_index()
```

```python
# calculate whiff_rate and OSwing_pct
data_swing_grouped['whiff_rate'] = data_swing_grouped['is_whiff']/data_swing_grouped['is_swing']
data_swing_grouped['OSwing_pct'] = data_swing_grouped['is_OSwing']/data_swing_grouped['is_outside']

data_swing_grouped = data_swing_grouped[['pitcher', 'whiff_rate', 'OSwing_pct']].copy()
```

```python
data_swing_grouped.head()
```

# Production Stats
### K/9

```python
# get full dataset
data_stats = data_train_full.copy()
```

```python
data_stats_pa = data_stats.groupby(['game_date', 'game_pk', 'at_bat_number']).agg({'game_year':max,
                                                                                        'home_team':max,
                                                                                        'away_team':max,
                                                                                        'player_name':max,
                                                                                        'batter':max,
                                                                                        'pitcher':max,
                                                                                        'inning_topbot':max,
                                                                                        'inning':max,
                                                                                        'events':'first'}).copy()
data_stats_pa = data_stats_pa.reset_index()
```

```python
# check NA
data_stats_pa['events'].isna().sum()
```

```python
# drop NA
data_stats_pa = data_stats_pa[data_stats_pa['events'].notna()].copy()
```

```python
is_K = ['strikeout', 'strikeout_double_play']
is_hit = ['single', 'home_run', 'double', 'triple']
is_BB = ['walk', 'hit_by_pitch']
is_HR = ['home_run']

data_stats_pa['is_K'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_K), 'is_K'] = 1

data_stats_pa['is_hit'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_hit), 'is_hit'] = 1

data_stats_pa['is_BB'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_BB), 'is_BB'] = 1

data_stats_pa['is_HR'] = 0
data_stats_pa.loc[data_stats_pa['events'].isin(is_HR), 'is_HR'] = 1

data_stats_pa['PA_pitched'] = 1
```

```python
# calculate outs generated
is_1out = ['field_out', 'strikeout', 'fielders_choice', 'sac_bunt', 'sac_fly', 'force_out',
          'fielders_choice_out', 'caught_stealing_2b', 'caught_stealing_home', 'pickoff_1b',
          'other_out', 'caught_stealing_3b', 'pickoff_2b', 'pickoff_caught_stealing_home',
          'pickoff_caught_stealing_3b', 'pickoff_3b', 'pickoff_caught_stealing_2b']
is_2out = ['grounded_into_double_play', 'double_play', 'strikeout_double_play',
          'sac_bunt_double_play', 'sac_fly_double_play', 'runner_double_play']
is_3out = ['triple_play']
is_0out = ['single', 'home_run', 'hit_by_pitch', 'walk', 'double', 'field_error', 'triple',
           'wild_pitch', 'catcher_interf', 'game_advisory', 'passed_ball', 'stolen_base_2b']
```

```python
data_stats_pa['num_outs'] = 3
data_stats_pa.loc[data_stats_pa.events.isin(is_1out), 'num_outs'] = 1
data_stats_pa.loc[data_stats_pa.events.isin(is_2out), 'num_outs'] = 2
data_stats_pa.loc[data_stats_pa.events.isin(is_0out), 'num_outs'] = 0
```

```python
# group data
data_stats_grouped = data_stats_pa.groupby(['pitcher']).agg({'is_K':sum, 'is_hit':sum, 
                                                            'is_BB':sum, 'is_HR':sum,
                                                            'num_outs':sum})
data_stats_grouped = data_stats_grouped.reset_index()
```

```python
data_stats_grouped['IP'] = data_stats_grouped['num_outs']/3
```

```python
data_stats_grouped['K_9'] = (data_stats_grouped['is_K']/data_stats_grouped['IP'])*9
data_stats_grouped['hit_9'] = (data_stats_grouped['is_hit']/data_stats_grouped['IP'])*9
data_stats_grouped['BB_9'] = (data_stats_grouped['is_BB']/data_stats_grouped['IP'])*9
data_stats_grouped['HR_9'] = (data_stats_grouped['is_HR']/data_stats_grouped['IP'])*9
```

```python
# save a copy of full columns
data_stats_grouped_full = data_stats_grouped.copy()
```

```python
data_stats_grouped = data_stats_grouped_full[['pitcher', 'IP','K_9', 'hit_9', 'BB_9', 'HR_9']].copy()
```
